```python
import numpy as np
import pandas as pd # Для работы с данными
import scipy.stats # При работе со статистикой
import matplotlib.pyplot as plt  # Библиотека для визуализации результатов 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('../Классификация/adult.csv')
data.head()
```

```python
data.info()
```

## Удаляем строки с незаполненными данными

```python
data[data['workclass'] == '?']
```

```python
data = data.loc[data['workclass'] != '?']
data.head()
```

## One Hot Encoding

```python
X = pd.get_dummies (data, columns = ['workclass', 'education', 'marital-status', 'occupation',
                                    'relationship', 'race', 'gender', 'native-country'])
```

```python
del X['income']
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit( data['income'])
le.classes_
```

```python
y = pd.Series (data = le.transform(data['income']))
y.head()
```

## Логистическая регрессия 

```python
Log_mod = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
Log_mod.fit(X_train, y_train)
predictions = Log_mod.predict(X_test)
```

```python
print ('Score на тренировочных данных')
Log_mod.score(X_train, y_train) 
```

```python
print ('Score на тестовых данных')
Log_mod.score(X_test,y_test)
```

## SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto') 
clf.fit(X_train, y_train)
```

```python
print ('Score на тренировочных данных')
clf.score(X_train, y_train) 
```

```python
print ('Score на тестовых данных')
clf.score(X_test,y_test)
```

## SVM со StandardScaler

```python
from sklearn.pipeline import make_pipeline # используем пайплайны для удобства
from sklearn.preprocessing import StandardScaler
clf = make_pipeline(StandardScaler(), SVC(gamma='auto')) 
clf.fit(X_train, y_train)
```

```python
print ('Score на тренировочных данных')
clf.score(X_train, y_train) 
```

```python
print ('Score на тестовых данных')
clf.score(X_test,y_test)
```
